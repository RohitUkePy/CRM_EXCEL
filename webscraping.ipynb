```python
import requests
from bs4  import BeautifulSoup
import pandas as pd

a= "https://www.indianindustriesdirectory.com/state/StateData/maharashtra.html"


r= requests.get(a)

soup = BeautifulSoup(r.text , "lxml")

na = soup.find_all("div", class_ ="result_container--right-location")
print(len(na))
name=[]
for i in na:
    aaa = i.text
    name.append(aaa)
df = pd.DataFrame(columns=name)

print (name)
```
